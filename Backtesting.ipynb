```python
import datetime as dt
import yfinance as yf
import pandas_datareader as web
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA
```

# Basic Backtesting Strategy: Moving Average

```python
class my_sma_strategy(Strategy):
  # this 'init' is different than '__init__'. 'init' is specifically identified by the Backtesting module
  def init(self):
    price = self.data.Close
    self.ma1 = self.I(SMA, price, 10) #this is giving us the 10 day moving average
    self.ma2 = self.I(SMA, price, 20) #this is giving us the 20 day moving average

  def next(self):
    if crossover(self.ma1, self.ma2):
      self.buy()
    elif crossover(self.ma2, self.ma1):
      self.sell()
```

```python
start = dt.date(2020, 1, 1)
end = dt.date(2022, 1, 1)

start = start.strftime("%Y-%m-%d")
end  = end.strftime("%Y-%m-%d")

data = yf.download("TSLA", start, end)

backtest = Backtest(data, my_sma_strategy, commission=.002, exclusive_orders=True)
stats = backtest.run()
print(stats)
```

```python
backtest.plot()
#triangles pointing up means that we bought
#triangles pointing down means that we sold
#the color greens means we profited on the action
#the color red means that we lost money on the action
```
